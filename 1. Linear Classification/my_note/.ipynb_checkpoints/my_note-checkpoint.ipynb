#### # First Classifier: Nearest Neighbor!

---
* Training Process
    - Memorize all training images (data and labels)
    - (easy)
* Testing Process
    - Use L1 or L2 (or other distance choices) distance to each and every training image, return label of the closest one
    - L1 : Manhattan Distance
    - L2 : Euclidean Distance
* Training Speed?
    - O(1)
* Testing Speed?
    - O(N)
    - Linearly slower according to the size of training data
* Contradict with what we want: efficient at test time
* Neural Net flip this: expensive at training, cheap test evaluation (almost constant)

### # Data Preproccesing
* Zero-centered data : Subtract the mean
* Normalize : divide with standard deviation
* Go further : PCA, Whitening
* **Careful** : any preprocessing statistics (e.g. the data mean) must only be computed on the training data, and then applied to the validation / test data

<img src="assets/data_preprocessing.PNG" style="width:500px"/>

#### ## Data Preprocessing for Images

* Center only for all train data
* Substract the mean image (e.g. AlexNet)
    - Mean Image = array[h, w, d]
* Substract per-channel mean (e.g. VGGNet)
    - Mean along each channel = 3 number

### # Logistic regression

---
<img src="assets/logistic_regression.PNG" style="width:150px"/>

* Weighted Sum to determine YES(1) or NO(0)
    - Receive input from outside
    - Multiplying by the weight connected to it
    - Sum up all of them
    - Limit output in the range [0-1]
* If output total high, then class = close to YES!
* Weight is a feature identifier
    - a value that state the importance of the atribute assosiated with it
    - if its input is important (to class 1), increase it
